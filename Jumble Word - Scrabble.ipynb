```python
import pandas as pd
import numpy as np
```

```python
ospd = pd.read_csv('ospd.txt')
ospd = ospd[(ospd['Word'].str.len() == 9)].reset_index().drop('index',axis=1)
```

```python
ospd.head()
```

```python
import inflect
p = inflect.engine()
word_list = list(filter(lambda item: p.singular_noun(item) == False,ospd['Word']))
```

```python
word_list[:5]
```
